```python
def view(log):
    f = open(path+log, "r")
    lines = f.readlines()
    ilp_line = lines[1]
    lp_line = lines[3]
    ilp_c = float(ilp_line.split()[-1])
    lp_c = float(lp_line.split()[-1])
    f.close()
    return (ilp_c == lp_c)
```

```python
import os
import re
path = "../log/"
logs = os.listdir(path)
tc = 0
fc = 0
for log in logs:
    if view(log):
        tc+=1
    else:
        f = open(path+log, "r")
        lines = f.readlines()
        for line in lines[5:]:
            line = re.sub(r'[^\w]', ' ', line)
            parts = line.split(" ")
            parts = [int(each) for each in parts if each != '']
            if len(parts) == 0:
                print()
                continue
            if(parts[2] != 0 or parts[3] != 0):
                print(line)
        print()
print(tc, fc)
```
